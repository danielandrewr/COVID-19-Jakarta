```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('./Dataset Analisis.xlsx')
```

```python
df.corr()
```

```python
x_ = df['Tanggal']
y_ = df['Meninggal']
y2_ = df['Vaksinasi 2']
```

```python
plt.figure(figsize=(16, 6))
sns.scatterplot(data=df, x=x_, y=y_)
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(data=df, x=x_, y=y2_)
```

```python
y = df['Meninggal'].values.reshape(-1, 1)
x = df['Vaksinasi 2'].values.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Make train and test model and Output line equation

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2)

linear_regres = LinearRegression()
linear_regres.fit(train_x, train_y)

sign = ""
if linear_regres.intercept_ >= 0:
    sign = "+"
else:
    sign = "-"

print("y = {} {} {}".format(linear_regres.coef_, sign, linear_regres.intercept_))
```

```python
# Show linear regres

predict_y = linear_regres.predict(test_x)
plt.scatter(test_x, test_y)
plt.plot(test_x, predict_y, c='g')

plt.xlabel('Meninggal')
plt.ylabel('Vaksinasi 2')

plt.title('Meninggal vs Vaksinasi 2')
```

```python
model_accuracy = 0
for i in range(1, 101):
    accuracy = linear_regres.score(test_x, test_y)
    model_accuracy += accuracy

#model_accuracy = model_accuracy / 100
print("Model Accuracy: {}%".format(model_accuracy))
```
