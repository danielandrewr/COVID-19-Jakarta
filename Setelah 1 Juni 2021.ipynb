```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df2 = pd.read_excel('./Dataset Analisis.xlsx')
```

```python
x2 = df2['Vaksinasi 2'].values.reshape(-1, 1)
y2 = df2['Positif Harian'].values.reshape(-1, 1)
```

```python
train_x2, test_x2, train_y2, test_y2 = train_test_split(x2, y2, test_size=0.2)

linear_regres = LinearRegression()
linear_regres.fit(train_x2, train_y2)

sign = ""
if linear_regres.intercept_ >= 0:
    sign = "+"
else:
    sign = "-"

print("y = {}x {} {}".format(linear_regres.coef_, sign, linear_regres.intercept_))
```

```python
predict_y2 = linear_regres.predict(test_x2)
plt.scatter(test_x2, test_y2)
plt.plot(test_x2, predict_y2, c='r')

plt.xlabel('Vaksinasi 2')
plt.ylabel('Positif Harian')

plt.title('Vaksinasi 2 vs Positif Harian Setelah 1 Juni 2021')
```
